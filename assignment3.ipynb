```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("student_info.csv")

plt.hist(df['study_hours'], bins=10, color='red', edgecolor='black')
plt.title('Distribution of Study Hours')
plt.xlabel('Hours')
plt.ylabel('Number of Students')
plt.grid(True)
plt.show()
gender_counts = df['gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Gender Distribution')
plt.axis('equal')  
plt.show()
avg_scores = df[['math_score', 'reading_score', 'writing_score']].mean()
plt.bar(avg_scores.index, avg_scores.values, color=['yellow', 'purple', 'black'])
plt.title('Average Scores by Subject')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100)
plt.grid(True, axis='y')
plt.show()
result_counts = df['final_result'].value_counts()
plt.pie(result_counts, labels=result_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgray'])
plt.title('Final Result Distribution')
plt.axis('equal')
plt.show()
parent_edu_counts = df['parent_education'].value_counts()
plt.bar(parent_edu_counts.index, parent_edu_counts.values, color='purple')
plt.title('Parent Education Level')
plt.xticks(rotation=30)
plt.xlabel('Education Level')
plt.ylabel('Number of Students')
plt.grid(True, axis='y')
plt.show()
plt.scatter(df['study_hours'], df['math_score'], alpha=0.7, color='blue')
plt.title('Study Hours vs Math Score')
plt.xlabel('Study Hours')
plt.ylabel('Math Score')
plt.grid(True)
plt.show()
colors = {'Pass': 'green', 'Fail': 'red'}
plt.scatter(df['attendance_rate'], df['study_hours'],
            c=df['final_result'].map(colors), alpha=0.6)
plt.title('Attendance Rate vs Study Hours (Colored by Final Result)')
plt.xlabel('Attendance Rate (%)')
plt.ylabel('Study Hours')
plt.grid(True)
plt.show()
```
